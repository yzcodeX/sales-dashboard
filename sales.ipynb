```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/sales_dataset.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

UnitPrice hat negative Werte

```python
# anzahl duplikate
df[df.duplicated()]
```

```python
# duplikate entfernen
df = df.drop_duplicates()
df[df.duplicated()]
```

```python
# zeige negative Werte an
cols_to_check = ["UnitPrice", "Revenue", "Discount", "FinalPrice"]
neg_values = df[cols_to_check] < 0

rows_with_neg = df[neg_values.any(axis=1)]
rows_with_neg
```

```python
# mit drop: drop erwartet ein index daher braucht man .index
# df = df.drop(df[df['UnitPrice'] < 0].index)

# behalte nur Zeilen, beidem UnitPrice positiv ist
df = df[df['UnitPrice'] >= 0]
df[df['UnitPrice'] < 0]
```

```python
# null werte prÃ¼fen
df.isnull().sum()
```

Null Werte bereinigen

```python
# anzeigen welche Zeilen Null sind
df[df.isna().any(axis=1)]
```

```python
# Zeilen entfernen die null sind, da es nur 2 Zeilen sind
df = df.dropna()
df[df.isna().any(axis=1)]
```

```python
df.isnull().sum()
```

```python
# index zurÃ¼cksetzen, da DatensÃ¤tze entfernt wurden
df = df.reset_index(drop = True)
df
```

```python
# eindeutige werte prÃ¼fen
df.nunique()
```

Region hat 7 eindeutige Werte. Das kann nicht sein, da es nur 4 Himmelsrichtungen gibt.

```python
# Inkonsistenzen bei Strings prÃ¼fen 
feature_list_string = ["Category", "Region", "SalesChannel", "Product"]

def check_unique_values():
    for col in feature_list_string:
        print(f"Einzigartige Werte in {col}: ", df[col].unique())
        
check_unique_values()
```

Region: falsche Schreibweise, fehlerhafte Werte bereinigen -> DatensÃ¤tze die Fehler enthalten durch den konkreten Wert ersetzen z.B. Weest -> West

```python
# falsche Schreibweise bereinigen
df["Region"] = df["Region"].replace({
    "Weest": "West",
    "N0rth": "North",
    "Soth": "South"
})
check_unique_values()
```

```python
df.info()
```

```python
# Daten wurden bereinigt und sind sauber, bereinigter Datensatz wird abgbespeichert
df.to_csv("datasets/sales_dataset_cleaned.csv", index = False)
```
